```python
import pandas as pd
```

```python
data = pd.read_csv("inputData.csv")
data.head(3)
```

```python
df = data.copy()

df[['Date', 'Flight Number', 'From-To', 'Class', 'Price']] = df['Flight Details'].str.split('//', expand=True)

df.head()
```

```python
df[['From', 'To']] = df['From-To'].str.split('-', expand=True)
```

```python
df.head()
```

```python
df = df.drop("Flight Details", axis = 1)
df.head(2)
```

```python
df.rename(columns = {'Flow Card?' : 'Flow Card'}, inplace = True)
```

```python
df = df.drop('From-To', axis = 1)
df.head()
```

```python
df.columns
```

```python
df = df[['Date', 'Flight Number', 'From', 'To', 'Class', 'Price', 'Flow Card', 'Bags Checked', 'Meal Type']]
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
df['Price'] = df['Price'].astype(float)
```

```python
df.info()
```

```python
df['Flow Card'] = df['Flow Card'].replace({0: 'No', 1: 'Yes'})
```

```python
df.head()
```

```python
FlowCardHloders = df[df['Flow Card'] == 'Yes']

NoFlowCardHloders = df[df['Flow Card'] == 'No']
```

```python
FlowCardHloders.head()
```

```python
NoFlowCardHloders.head()
```

```python
print(f"There's {FlowCardHloders.shape[0]} Flow card holder and {NoFlowCardHloders.shape[0]} No flow card holder")
```

```python
FlowCardHloders.to_csv('FlowCardHolders.csv', index=False)

NoFlowCardHloders.to_csv('NoFlowCardHolders.csv', index=False)
```
